```r
df <- read.csv(file="data_020320.csv",sep=",",header=TRUE)
```

```r
colnames(df)
dim(df)
```

```r
colnames(df)[2] = '1st_Term_Funding'
colnames(df)[3] = '2nd_Term_Funding'
colnames(df)[4] = '3rd_Term_Funding'
colnames(df)[5] = '4th_Term_Funding'
colnames(df)[12] = 'funding_indicator'
colnames(df)[17] = 'net_cost'
colnames(df)[34] = 'GPA_indicator'
colnames(df)[35] = 'GPA_trend'
colnames(df)[37] = 'Average_GPA'
colnames(df)[38] = 'academic_period'
colnames(df)
dim(df)
```

```r
lc = levels(df$Citizenship)
as.matrix(lc)
```

```r
df$Citizenship[which(df$Citizenship=='Afria')]=lc[2]
df$Citizenship = droplevels(df$Citizenship)
as.matrix(levels(df$Citizenship))
```

```r
write.csv(df, file = "data_020320_cleaned")
```

```r
df$Citizenship[which(df$Citizenship=='Asian')] = "American Asian"


lc[4] = 'Asian American'
lc[8] = 'Hawaiian or Other Pacific Islander American'
lc[9] = 'Hispanic or Latino American'
lc[14] = 'White American'
```
